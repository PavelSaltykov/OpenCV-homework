```python
import cv2
from IPython.display import display
from PIL import Image


def display_image(img, conversion=cv2.COLOR_BGR2RGB):
    if conversion:
        img = cv2.cvtColor(img, conversion)
    display(Image.fromarray(img))


def get_frame_from_camera(device_index=0):
    cap = cv2.VideoCapture(device_index)
    
    while True:
        _, frame = cap.read()
        cv2.imshow("Camera", frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()
    return frame
```

## Original Image

```python
image_path = 'rubberwhale.png'
original_image = cv2.imread(image_path)
display_image(original_image)
```

## Task 1
 canny edges  .

```python
edges = cv2.Canny(original_image, 150, 200)
display_image(edges)
```

## Task 2
   grayscale.

```python
gray = cv2.cvtColor(original_image, cv2.COLOR_BGR2GRAY)
display_image(gray)
```

## Task 3
   hsv.

```python
hsv = cv2.cvtColor(original_image, cv2.COLOR_BGR2HSV)
display_image(hsv)
```

## Task 4
    .

```python
flipped = cv2.flip(original_image, 0)
display_image(flipped)
```

## Task 5
   45 .

```python
h, w, d = original_image.shape
center = (w // 2, h // 2)
M = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated = cv2.warpAffine(original_image, M, (w, h))
display_image(rotated)
```

## Task 6
  .

```python
brightened = cv2.convertScaleAbs(original_image, alpha=1, beta=100)
display_image(brightened)
```

## Task 7
  .

```python
contrasted = cv2.convertScaleAbs(original_image, alpha=2.0, beta=0)
display_image(contrasted)
```

## Task 8
  .

```python
_, threshold_image = cv2.threshold(gray, 127, 255, 0)
display_image(threshold_image)
```
